# Cleaning Zomato Reviews Data for Visualization
[Data Source](https://www.kaggle.com/datasets/prepinstaprime/tomato-reviews)

```python
# Importing Required Libraries
import pandas as pd
```

```python
raw_data = pd.read_csv("zomato.csv")
```

```python
raw_data.info()
```

```python
raw_data.head()
```

### Dropping Columns
Dropping Columns that i will not be using or can be used for visualization

```python
raw_data.drop(["url","address","phone","dish_liked","reviews_list","menu_item"],axis=1,inplace=True)
```

```python
raw_data.head()
```

### Renaming Columns 
Renaming Columns so they become easy to use and make more sence 

```python
raw_data.rename(
    columns={
        'name':'restaurant',
        'rate':'rating',
        'cuisines':'food_type',
        'approx_cost(for two people)':'price',
        'listed_in(type)':'type',
        'listed_in(city)':'city'
    },
    inplace=True)
```

```python
raw_data.head(20)
```

`Looking at Null Values`

```python
raw_data.isna().sum()
```

```python
raw_data.isna().sum().sum()
```

```python
len(raw_data)
```

```python
# Removing Null Values
raw_data.dropna(inplace = True)
```

```python
len(raw_data)
```

Length (__Number of Rows__) of our data before and after dropping __Null Values__

```python
raw_data.head()
```

#### Dropping Duplicates

```python
raw_data.duplicated().value_counts()
```

```python
# dropped data 
raw_data.drop_duplicates(keep='last',inplace=True)
```

```python
len(raw_data)
```

### Now we have done all necessary data cleaning we can now proceed to data visualization

```python
# Saved our Pandas dataframe in System for Visualization
raw_data.reset_index(drop=True,inplace=True)
raw_data.to_excel("zomato_cleaned_data.xlsx")
```
